TASK 1: Data Information aur Basic Stats **Check**

```python
"""
📌 TASK 1.1: Data Information aur Basic Statistics Check
-----------------------------------------------------------
KYA: Dataset ki basic information dekh rahe hain
KYUN: Ye check karne ke liye ki:
      - Kitne rows aur columns hain
      - Kaunsa column kis type ka hai (number/text)
      - Koi missing value to nahi
      - Data ki range kya hai (min, max, average)
KAUNSA KAAM: Data Understanding (Pehla step)
"""

print("="*50)
print("📊 DATA INFORMATION")
print("="*50)

# Check data types and missing values
df.info()

print("\n" + "="*50)
print("📈 BASIC STATISTICS (Numerical Columns)")
print("="*50)
df.describe()

print("\n" + "="*50)
print("❓ MISSING VALUES CHECK")
print("="*50)
df.isnull().sum()
```

```python
"""
📌 TASK 1.3: ID Column Remove karna
-----------------------------------------------------------
KYA: 'id' column ko hata rahe hain
KYUN: ID column se koi prediction nahi hoti, ye sirf unique identifier hai
      Model training mein iski koi zaroorat nahi
KAUNSA KAAM: Feature Selection (Unnecessary columns hata rahe hain)
"""

print("="*50)
print("🆔 ID COLUMN REMOVAL")
print("="*50)

if 'id' in df.columns:
    df = df.drop('id', axis=1)
    print("✅ 'id' column removed successfully")

    # Check kitne columns bache
    print(f"\n📌 Columns now: {list(df.columns)}")
    print(f"📊 Total columns: {len(df.columns)}")
else:
    print("ℹ️ 'id' column already removed")
    print(f"\n📌 Current columns: {list(df.columns)}")

# Dataset ka size check
print(f"\n📈 Final dataset size: {len(df)} rows × {len(df.columns)} columns")
```

```python
"""
📌 TASK 1.2: Duplicate Rows Check aur Remove
-----------------------------------------------------------
KYA: Duplicate (ek jaisi) rows check kar rahe hain
KYUN: Machine learning ke liye duplicate data galat results de sakta hai
      Isliye duplicate rows ko remove karna zaroori hai
KAUNSA KAAM: Data Cleaning
"""

print("="*50)
print("🔄 DUPLICATE ROWS CHECK")
print("="*50)

duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")

if duplicates > 0:
    df = df.drop_duplicates()
    print(f"✅ After removing duplicates: {len(df)} rows remain")

    # Check kitne rows hata diye
    removed = 630000 - len(df)
    print(f"🗑️ Removed {removed} duplicate rows")
else:
    print("✅ No duplicate rows found")

print(f"\n📊 Current dataset size: {len(df)} rows")
```

```python
"""
📌 TASK 1.4: Column Types Identify karna
-----------------------------------------------------------
KYA: Kaunse columns categorical (text) hain aur kaunse numerical (numbers)
KYUN: Encoding ke liye pata hona chahiye ki kaunse columns ko convert karna hai
      Categorical columns ko machine readable banana padta hai
KAUNSA KAAM: Data Preparation (Encoding ke liye ready hona)
"""

print("="*50)
print("🔤 CATEGORICAL & NUMERICAL COLUMNS IDENTIFICATION")
print("="*50)

# Categorical columns (object type)
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Numerical columns (int64, float64 type)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("\n📝 CATEGORICAL COLUMNS (Need Encoding):")
print("-" * 40)
for i, col in enumerate(categorical_cols, 1):
    unique_values = df[col].nunique()
    print(f"{i}. {col} - {unique_values} unique values")

print("\n🔢 NUMERICAL COLUMNS (Already numbers):")
print("-" * 40)
for i, col in enumerate(numerical_cols, 1):
    print(f"{i}. {col}")

print("\n" + "="*50)
print(f"✅ Total Categorical Columns: {len(categorical_cols)} (Need Encoding)")
print(f"✅ Total Numerical Columns: {len(numerical_cols)} (Ready for model)")
print(f"📊 Total Columns: {len(df.columns)}")

# Show sample data for categorical columns
print("\n" + "="*50)
print("📋 SAMPLE DATA - First 3 rows of categorical columns:")
print("="*50)
print(df[categorical_cols].head(3))
```

TASK 2 START: ENCODING CATEGORICAL VARIABLES
Ab hum Encoding karenge. Pehle decide karte hain kaunsa encoding kahan use karna hai:

Encoding Strategy:
Label Encoding - Jahan 2-3 unique values hain (ordered categories)

One-Hot Encoding - Jahan 4+ unique values hain (no order)

```python
"""
📌 TASK 2.1: Label Encoding
-----------------------------------------------------------
KYA: Categorical columns ko numbers mein convert kar rahe hain
      (gender, internet_access, sleep_quality, facility_rating, exam_difficulty)
KYUN: Machine learning models sirf numbers samajhte hain, text nahi
KAUNSA KAAM: Feature Encoding (Pehla encoding step)
"""

from sklearn.preprocessing import LabelEncoder
import pickle

print("="*50)
print("🏷️ LABEL ENCODING")
print("="*50)

# Dictionary to store encoders for later use
encoders = {}
df_encoded = df.copy()

# Columns for Label Encoding
label_encode_cols = ['gender', 'internet_access', 'sleep_quality',
                     'facility_rating', 'exam_difficulty']

print("\n📝 Encoding these columns with LabelEncoder:")
print("-" * 40)

for col in label_encode_cols:
    # Create and fit encoder
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df[col])

    # Store encoder
    encoders[f'label_{col}'] = le

    # Print mapping
    print(f"\n{col.upper()}:")
    print(f"  Unique values: {df[col].unique().tolist()}")
    print(f"  Encoded as: {le.transform(le.classes_).tolist()}")
    print(f"  Mapping: {dict(zip(le.classes_, le.transform(le.classes_)))}")

print("\n" + "="*50)
print("✅ Label Encoding Complete")
print("="*50)

# Show sample after encoding
print("\n📋 Sample data after Label Encoding (first 5 rows):")
print(df_encoded[label_encode_cols].head())
```

CELL 2: One-Hot Encoding

```python
"""
📌 TASK 2.2: One-Hot Encoding
-----------------------------------------------------------
KYA: Bache hue categorical columns ko One-Hot Encoding se convert kar rahe hain
      (course, study_method, student_performance)
KYUN: In columns mein unique values zyada hain (4+), isliye One-Hot Encoding better hai
      Har unique value ke liye naya column banega
KAUNSA KAAM: Feature Encoding (Dusra encoding step)
"""

print("="*50)
print("🔄 ONE-HOT ENCODING")
print("="*50)

# Columns for One-Hot Encoding
onehot_encode_cols = ['course', 'study_method', 'student_performance']

print(f"\n📝 Encoding these columns with One-Hot Encoding:")
print(f"   {onehot_encode_cols}")
print("-" * 50)

# Check unique values before encoding
for col in onehot_encode_cols:
    unique_vals = df_encoded[col].unique()
    print(f"\n{col.upper()}:")
    print(f"  Unique values: {unique_vals}")
    print(f"  Total categories: {len(unique_vals)}")

# Apply One-Hot Encoding
print("\n" + "="*50)
print("Applying One-Hot Encoding...")
print("="*50)

# Get dummies (One-Hot Encoding)
df_encoded = pd.get_dummies(df_encoded, columns=onehot_encode_cols, prefix=onehot_encode_cols)

print("\n✅ One-Hot Encoding Complete!")

# Show new columns created
new_cols = [col for col in df_encoded.columns if any(x in col for x in onehot_encode_cols)]
print(f"\n📊 New columns created ({len(new_cols)} total):")
print("-" * 50)
for col in new_cols[:10]:  # Show first 10
    print(f"  • {col}")

print("\n" + "="*50)
print(f"📈 Dataset shape after encoding: {df_encoded.shape}")
print(f"   Rows: {df_encoded.shape[0]:,}")
print(f"   Columns: {df_encoded.shape[1]}")
print("="*50)

# Show sample of encoded data
print("\n📋 Sample data after One-Hot Encoding (first 3 rows, selected columns):")
sample_cols = ['gender', 'age', 'study_hours'] + new_cols[:5]
print(df_encoded[sample_cols].head(3))
```

TASK 3: FEATURE SCALING
Ab numerical columns ko scale karenge.

```python
"""
📌 TASK 3: Feature Scaling (StandardScaler)
-----------------------------------------------------------
KYA: Numerical columns ko scale kar rahe hain (0 ke around mean, 1 standard deviation)
KYUN: Different scales (e.g., age 18-24, study_hours 0-10) model ko confuse karte hain
      Scaling se sab columns ko equal importance milti hai
KAUNSA KAAM: Feature Scaling (Model ke liye data prepare)
"""

from sklearn.preprocessing import StandardScaler
import numpy as np

print("="*50)
print("📏 FEATURE SCALING (STANDARDSCALER)")
print("="*50)

# Numerical columns to scale
numerical_cols = ['age', 'study_hours', 'class_attendance', 'sleep_hours', 'exam_score']

print(f"\n📝 Scaling these numerical columns:")
print(f"   {numerical_cols}")
print("-" * 50)

# Show statistics before scaling
print("\n📊 Statistics BEFORE Scaling:")
print("-" * 30)
for col in numerical_cols:
    print(f"{col}:")
    print(f"  Mean = {df_encoded[col].mean():.2f}")
    print(f"  Std  = {df_encoded[col].std():.2f}")
    print(f"  Min  = {df_encoded[col].min():.2f}")
    print(f"  Max  = {df_encoded[col].max():.2f}")

# Initialize and fit scaler
scaler = StandardScaler()
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

# Store scaler for later use
encoders['scaler'] = scaler

print("\n" + "="*50)
print("✅ Scaling Complete!")
print("="*50)

# Show statistics after scaling
print("\n📊 Statistics AFTER Scaling:")
print("-" * 30)
for col in numerical_cols:
    print(f"{col}:")
    print(f"  Mean = {df_encoded[col].mean():.2f}")
    print(f"  Std  = {df_encoded[col].std():.2f}")
    print(f"  Min  = {df_encoded[col].min():.2f}")
    print(f"  Max  = {df_encoded[col].max():.2f}")

# Show sample of scaled data
print("\n" + "="*50)
print("📋 Sample data after Scaling (first 3 rows):")
print("-" * 50)
print(df_encoded[numerical_cols].head(3))

print("\n" + "="*50)
print(f"📈 Final dataset shape: {df_encoded.shape}")
print(f"   Rows: {df_encoded.shape[0]:,}")
print(f"   Columns: {df_encoded.shape[1]}")
print("="*50)
```

TASK 4: TRAIN-TEST SPLIT
Ab data ko training (80%) aur testing (20%) mein divide karte hain.

```python
"""
📌 TASK 4: Train-Test Split
-----------------------------------------------------------
KYA: Data ko do parts mein divide kar rahe hain:
     - Training Set (80%): Model ko train karne ke liye
     - Testing Set (20%): Model ko test karne ke liye
KYUN: Model ko unseen data par evaluate karna hota hai
      Alag data par test karenge to pata chalega ki model kitna accurate hai
KAUNSA KAAM: Final Data Preparation (Model ke liye ready)
"""

from sklearn.model_selection import train_test_split
import pickle

print("="*50)
print("✂️ TRAIN-TEST SPLIT")
print("="*50)

# Step 1: Separate Features (X) and Target (y)
# Target variable kya hai? - 'student_performance' (One-Hot encoded columns)
target_cols = [col for col in df_encoded.columns if col.startswith('student_performance_')]
print(f"\n🎯 Target columns (One-Hot encoded):")
for col in target_cols:
    print(f"   • {col}")

# For classification, we need single target column (not One-Hot)
# Convert One-Hot back to single column
target_original = df[['student_performance','exam_score']]  # Original target from before encoding
X = df_encoded.drop(columns=target_cols + ['exam_score'])  # Features (remove target columns)

y = df['student_performance']  # Target

print(f"\n📊 Features (X) shape: {X.shape}")
print(f"📊 Target (y) shape: {y.shape}")
print(f"\n📝 Features columns: {list(X.columns)}")
print(f"\n🎯 Target unique values: {y.unique()}")

# Step 2: Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,           # 20% for testing
    random_state=42,          # Reproducible results
    stratify=y               # Maintain class distribution
)

print("\n" + "="*50)
print("✅ SPLIT COMPLETE")
print("="*50)
print(f"\n📈 Training set size (80%):")
print(f"   X_train: {X_train.shape}")
print(f"   y_train: {y_train.shape}")
print(f"   Samples: {len(X_train):,}")

print(f"\n📈 Testing set size (20%):")
print(f"   X_test: {X_test.shape}")
print(f"   y_test: {y_test.shape}")
print(f"   Samples: {len(X_test):,}")

# Step 3: Check distribution
print("\n" + "="*50)
print("📊 TARGET DISTRIBUTION")
print("="*50)

print("\n📊 Original dataset:")
print(y.value_counts(normalize=True).round(3) * 100)

print("\n📊 Training set:")
print(y_train.value_counts(normalize=True).round(3) * 100)

print("\n📊 Testing set:")
print(y_test.value_counts(normalize=True).round(3) * 100)

# Step 4: Save all files for Model Training team
print("\n" + "="*50)
print("💾 SAVING FILES FOR MODEL TRAINING TEAM")
print("="*50)

# Save datasets
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)

# Save encoders and scaler
with open('encoders.pkl', 'wb') as f:
    pickle.dump(encoders, f)

# Save feature names
with open('feature_names.pkl', 'wb') as f:
    pickle.dump(list(X.columns), f)

print("\n✅ Files saved successfully:")
print("   1. X_train.csv - Training features")
print("   2. X_test.csv - Testing features")
print("   3. y_train.csv - Training target")
print("   4. y_test.csv - Testing target")
print("   5. encoders.pkl - LabelEncoders and Scaler")
print("   6. feature_names.pkl - Feature names list")

print("\n" + "="*50)
print("🎉 TASK COMPLETE! ML-READY DATASET READY")
print("="*50)
print("\n📢 Message for Model Training Team (Ashish & Hardik):")
print("-" * 50)
print("Hi Team,")
print("Feature Engineering complete! Files ready for model training:")
print("• X_train.csv, X_test.csv - Features (26 columns)")
print("• y_train.csv, y_test.csv - Target (student_performance)")
print("• encoders.pkl - For Flask integration later")
print("\nDataset stats:")
print(f"• Training samples: {len(X_train):,}")
print(f"• Testing samples: {len(X_test):,}")
print(f"• Features: {X_train.shape[1]}")
print("\nTarget classes: Excellent, Good, Average, Poor")
print("\nPlease start model training! 🚀")
```
